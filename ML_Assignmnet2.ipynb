**Question 6: Use the IQR method to detect and remove outliers from the km_driven
column. Show before-and-after summary statistics.**


```python
#Answer 6:
import pandas as pd
import os

file_path = 'BIKE DETAILS (1).csv'


if not os.path.exists(file_path):
    print(f"Error: The file '{file_path}' was not found. Please upload it to your Colab environment and then re-run this cell.")
else:
    df = pd.read_csv(file_path)

    print("--- Before Removal ---")
    before_stats = df['km_driven'].describe()
    print(before_stats)

    Q1 = df['km_driven'].quantile(0.25)
    Q3 = df['km_driven'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_filtered = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

    print("\n--- After Removal ---")
    after_stats = df_filtered['km_driven'].describe()
    print(after_stats)
```

**Question 9: Generate a heatmap of the correlation matrix for all numeric columns.
What correlations stand out the most?**

```python
#Answer 9:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

file_path = 'BIKE DETAILS (1).csv'

# Check if the file exists (for standalone execution or if df is not yet in locals)
if not os.path.exists(file_path):
    print(f"Error: The file '{file_path}' was not found. Please upload it to your Colab environment and then re-run this cell.")
else:
    # Load the data if df is not already defined in the current session
    if 'df' not in locals():
        df = pd.read_csv(file_path)

    # Select numeric columns and calculate the correlation matrix
    numeric_df = df.select_dtypes(include=['int64', 'float64'])
    corr_matrix = numeric_df.corr()

    # Generate the heatmap
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Matrix of Numeric Columns')

    # Save the plot (optional)
    # plt.savefig('correlation_heatmap.png')
    plt.show()
```

**Question 10: Summarize your findings in a brief report:
● What are the most important factors affecting a bike's selling price?
● Mention any data cleaning or feature engineering you performed.**

```python
#Answer 10:

import pandas as pd
import numpy as np
import os

file_path = 'BIKE DETAILS (1).csv'


if not os.path.exists(file_path):
    print(f"Error: The file '{file_path}' was not found. Please upload it to your Colab environment and then re-run this cell.")
else:

    if 'df' not in locals():
        df = pd.read_csv(file_path)


    df['ex_showroom_price'] = df.groupby('name')['ex_showroom_price'].transform(lambda x: x.fillna(x.median()))

    df['ex_showroom_price'] = df['ex_showroom_price'].fillna(df['ex_showroom_price'].median())


    Q1 = df['km_driven'].quantile(0.25)
    Q3 = df['km_driven'].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR

    df_cleaned = df[df['km_driven'] <= upper_bound].copy()


    df_final = pd.get_dummies(df_cleaned, columns=['seller_type'])


    numeric_cols = df_final.select_dtypes(include=[np.number])
    correlations = numeric_cols.corr()['selling_price'].sort_values(ascending=False)

    print("--- Processed Data Summary ---")
    print(f"Rows remaining after cleaning: {len(df_final)}")
    print("\n--- Correlations with Selling Price ---")
    print(correlations)
    print("\n--- Final Data Sample ---")
    print(df_final.head())
```
